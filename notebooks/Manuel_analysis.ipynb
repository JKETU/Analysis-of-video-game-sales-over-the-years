```python
#to get the sales ranking
sorted_df = merged_df.sort_values(by='global_sales', ascending=False)
sorted_df
```

```python
# Top-selling games by Manuel.
#My balance : Wii sports is the name of the best-selling game.
top_selling_games = sorted_df.groupby('name')['global_sales'].sum().sort_values(ascending=False)
print("Top-selling games:\n", top_selling_games.head(10))

```

```python
# Sales by region
region_sales = sorted_df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum()
print("\nSales by region:\n", region_sales)
```

```python
# Extract the franchise from the 'name' column
# The game franchise is the first word in the 'name' column
sorted_df['Franchise'] = sorted_df['name'].str.split().str[0]
```

```python
sorted_df['Franchise']
```

```python
# Group by 'Franchise', 'Year', and 'Publisher', and sum the sales
#list1=['name', 'year', 'publisher']
#list2=['global_sales', 'na_sales', 'eu_sales']
franchise_success = sorted_df.groupby(['name', 'year', 'publisher'])[['global_sales', 'na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum().reset_index()
#franchise_success = sorted_df.groupby([list1])[list2].sum().reset_index()
display(franchise_success)
```

```python
# Visualize franchise success over time for the top 10 franchises by total sales
top_franchises = franchise_success.groupby('name')['global_sales'].sum().nlargest(10).index
```

```python
print(top_franchises)
```

```python
plt.figure(figsize=(15, 10))
for name in top_franchises:
    franchise_data = franchise_success[franchise_success['name'] == name]
    sns.countplot(x='name', data=franchise_data, label=name)
plt.title('Franchise Success over time (Top 10 Franchises)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
#plt.savefig("../video_game_analysis/figures/top_10_franchises_by_total_sales
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# df2 is a pandas source DataFrame and sorted_df, the modified df.
platform_sales = sorted_df.groupby('name').agg(number_of_platforms=('platform_y', 'count'), total_global_sales=('global_sales', 'sum')).reset_index()

platform_sales['platform_number'] = platform_sales['number_of_platforms'].apply(lambda x: '1 platform' if x == 1 else '2 platform' if x == 2 else '3 platform' if x == 3 else '4 platform' if x == 4 else '5 or more platform')

result = platform_sales.groupby('platform_number').agg(count_games=('number_of_platforms', 'count'), totalsales_per_number_of_platform=('total_global_sales', 'sum')).reset_index()

result = result.sort_values('totalsales_per_number_of_platform', ascending=False)
display(result)
```
