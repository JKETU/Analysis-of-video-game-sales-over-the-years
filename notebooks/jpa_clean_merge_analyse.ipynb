# Analysis as per Initial questions:
Critical Reception and User Ratings
Axis: Average critic score, user score, and review metrics.

Potential Analysis:
1. Correlation between critic and user scores: Investigate the relationship between critic reviews and user ratings, and identify any discrepancies.
2. Impact of reviews on sales: Explore how game ratings (both critic and user) impact sales.
3. Outliers: Identify games that have significantly different critic and user scores and try to understand the reasons.

```python
import plotly.express as px
import numpy as np
```

```python
# Convert columns to integers/float
filtered_df['meta_score'] = pd.to_numeric(filtered_df['meta_score'], errors='coerce')
filtered_df['user_review'] = pd.to_numeric(filtered_df['user_review'], errors='coerce')

# Drop any rows with NaN values from the conversion
filtered_df = filtered_df.dropna(subset=['meta_score', 'user_review'])

# Calculate correlation
correlation = np.corrcoef(filtered_df['meta_score'], filtered_df['user_review'])[0, 1]
print(f"Correlation between Critic Scores and User Scores: {correlation:.2f}")

# Create scatter plot
fig = px.scatter(filtered_df, x='meta_score', y='user_review',
                 labels={'meta_score': 'Critic Score', 'user_review': 'User Score'},
                 title='Correlation Between Critic and User Scores')

fig.show()
```

```python
# Pivot table for average critic and user scores by genre
pivot_genre_scores = filtered_df.pivot_table(values=['meta_score', 'user_review'],
                                             index='genre',
                                             aggfunc='mean').reset_index()

display(pivot_genre_scores)
```

```python
# Pivot table for global sales by platform and year
pivot_sales = filtered_df.pivot_table(values='global_sales',
                                      index='platform_x',
                                      columns='year',
                                      aggfunc='sum').fillna(0)

display(pivot_sales)
```

```python
# Create a pivot table to calculate the sum meta_score and user_review for each region
pivot_jp_df = filtered_df.pivot_table(
    index='jp_sales', 
    values=['meta_score', 'user_review'],
    aggfunc='count'
)
pivot_jp_df

pivot_na_df = filtered_df.pivot_table(
    index='na_sales', 
    values=['meta_score', 'user_review'],
    aggfunc='count'
)
pivot_na_df

pivot_eu_df = filtered_df.pivot_table(
    index='eu_sales', 
    values=['meta_score', 'user_review'],
    aggfunc='count'
)
pivot_eu_df

pivot_global_df = filtered_df.pivot_table(
    index='global_sales', 
    values=['meta_score', 'user_review'],
    aggfunc='count'
)
pivot_global_df
```

```python
import plotly.graph_objects as go

# Calculate the sum meta_score and user_review for each region
meta_scores = [
    pivot_na_df['meta_score'].count(),
    pivot_eu_df['meta_score'].count(),
    pivot_jp_df['meta_score'].count(),
    pivot_global_df['meta_score'].count()
]

user_reviews = [
    pivot_na_df['user_review'].count(),
    pivot_eu_df['user_review'].count(),
    pivot_jp_df['user_review'].count(),
    pivot_global_df['user_review'].count()
]

# List of regions
regions = ['North America', 'Europe', 'Japan', 'Global']

```

```python

# Plotting the sum meta_score by region
fig_meta = go.Figure()

fig_meta.add_trace(go.Bar(
    x=regions,
    y=meta_scores,
    name='Meta Score',
    marker_color='indianred'
))

fig_meta.update_layout(
    title="Count Meta Score by Region",
    xaxis_title="Region",
    yaxis_title="Meta Score",
    barmode='group'
)

fig_meta.show()
```

```python
# Plotting the sum user_review by region
fig_user = go.Figure()

fig_user.add_trace(go.Bar(
    x=regions,
    y=user_reviews,
    name='User Review',
    marker_color='lightsalmon'
))

fig_user.update_layout(
    title="Count User Review by Region",
    xaxis_title="Region",
    yaxis_title="User Review",
    barmode='group'
)

fig_user.show()
```
