```python
# Create a pivot table showing total global sales per genre per platform
genre_platform_sales = merged_df.pivot_table(index='genre', columns='platform_x', values='global_sales', aggfunc='sum', fill_value=0)

# Plot the pivot table as a heatmap to visualize the sales distribution
plt.figure(figsize=(14, 8))
sns.heatmap(genre_platform_sales, cmap='YlOrRd', annot=True, fmt='.2f')
plt.title('Global Sales by Genre Across Platforms')
plt.xlabel('Platform')
plt.ylabel('Genre')
plt.show()
```

```python

# Step 1: Aggregate Data by Genre and Platform
# You can choose to aggregate by the count of games or total global sales
# Here we'll do both as examples:

# 1. Aggregate by the count of games
genre_platform_count = merged_df.groupby(['genre', 'platform_x']).size().reset_index(name='game_count')

# 2. Aggregate by total global sales
genre_platform_sales = merged_df.groupby(['genre', 'platform_x'])['global_sales'].sum().reset_index()

# Step 2: Identify the Top Ten Combinations
# Sort by game count
top_ten_genre_platform_count = genre_platform_count.sort_values(by='game_count', ascending=False).head(10)

# Sort by global sales
top_ten_genre_platform_sales = genre_platform_sales.sort_values(by='global_sales', ascending=False).head(10)

# Step 3: Visualize the Results

# 1. Top Ten Genre-Platform Combinations by Game Count
plt.figure(figsize=(14, 7))
sns.barplot(x='game_count', y='platform_x', hue='genre', data=top_ten_genre_platform_count, palette='viridis')
plt.title('Top 10 Genre-Platform Combinations by Number of Games')
plt.xlabel('Number of Games')
plt.ylabel('Platform')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# 2. Top Ten Genre-Platform Combinations by Global Sales
plt.figure(figsize=(14, 7))
sns.barplot(x='global_sales', y='platform_x', hue='genre', data=top_ten_genre_platform_sales, palette='coolwarm')
plt.title('Top 10 Genre-Platform Combinations by Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Platform')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 3. Average Sales per Game per Platform
avg_sales_per_game = merged_df.groupby('platform_x')['global_sales'].mean().sort_values(ascending=False)

# Plotting average sales per game by platform
plt.figure(figsize=(14, 7))
avg_sales_per_game.plot(kind='bar', color='coral')
plt.title('Average Sales per Game by Platform')
plt.xlabel('Platform')
plt.ylabel('Average Sales per Game (in millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Distribution of Games Across Platforms Over Time
games_over_time = merged_df.groupby(['year', 'platform_y']).size().unstack().fillna(0)

# Plotting the distribution of games across platforms over time
plt.figure(figsize=(10, 5))
games_over_time.plot(kind='line', stacked=True, figsize=(10, 5), colormap='tab20')
plt.title('Distribution of Games Across Platforms Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Games Released')
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# 2. Sales Trends Across Platforms Over Time
sales_over_time = merged_df.groupby(['year', 'platform_y'])['global_sales'].sum().unstack().fillna(0)

# Plotting the sales trends across platforms over time
plt.figure(figsize=(14, 7))
sales_over_time.plot(kind='line', figsize=(14, 7), colormap='tab20')
plt.title('Sales Trends Across Platforms Over Time')
plt.xlabel('Year')
plt.ylabel('Total Global Sales (in millions)')
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
